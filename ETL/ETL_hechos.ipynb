```python
import pandas as pd

def CalcularNulos(df):
    for columna in df:
        nulos = 100*(df[columna].isnull().sum()/df.shape[0])
        porcentaje = "{:.1f}".format(nulos)
        print("el porcentaje de datos nulos en la columna ",df[columna].name," es del ",porcentaje,"%")

    print('-------------------------------------------------------------------------------------------')
```
